```python
def cvrplot(x):
    plt.figure(figsize=(10,10))
    
    name = x + ' Vs Converted'
    plt.title(name)
    
    
    chart = sns.lineplot(x = df[x],y=df['Converted'])
    chart.set_xticklabels(rotation=90, horizontalalignment='right')
    plt.savefig(name)
    
    
    
    
```

```python
def cvrscatplot(x):
    
    name = x + ' Vs Converted'
    plt.title(name)
    
    sns.scatterplot(data=df,x = df[x],y=df['Converted'])

    
    
    
    plt.savefig(name)
    
```
